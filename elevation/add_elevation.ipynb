```python
import srtm
import pandas as pd
import time
from requests.exceptions import SSLError
import numpy as np
import sys
```

```python
elevation_data = srtm.get_data(local_cache_dir="./")
```

```python
print('Elevation (meters):', elevation_data.get_elevation(50.0447236, 18.7006401))
```

```python
df = pd.read_csv("coordinates.csv")
```

```python
step = 1000000
size = df.shape[0]
iterations = int(size/step)
rest = iterations*step
counter = 0
```

```python
df.columns
```

```python
def add_elevation(row):
    try:
        row["elevation"] = elevation_data.get_elevation(row["lat"],row["lon"])
    except:
        e = sys.exc_info()[0]
        print(f"Error: {e} for coordinates: ({row['lat']},{row['lon']})")
        row["elevation"] = np.nan
    return row
```

```python
start = 29
```

```python
for i in range(start,iterations):
    print(f"Started iteration {i}")
    start = time.time()
    data = df[i*(step):(i+1)*step].copy()
    data = data.apply(add_elevation,axis=1)
    data.to_csv(f"elevation_{i}.csv",index=None)
    end = time.time()
    print(f"Finished iteration {i} in {end-start} seconds")
```

```python
print(f"Started iteration {iterations}")
start = time.time()
data = df[rest:].copy()
print(f"Number of records: {data.size}")
data = data.apply(add_elevation,axis=1)
data.to_csv(f"elevation_{iterations}.csv",index=None)
end = time.time()
print(f"Finished iteration {iterations} in {end-start} seconds")
```
