```python
import pandas as pd
```

```python
number_of_files = 43
```

```python
df = pd.DataFrame()
```

```python
for i in range(number_of_files):
    print(f"File number: {i}")
    edf = pd.read_csv(f"elevation_{i}.csv")
    df = pd.concat([df,edf])
```

```python
df.reset_index(inplace=True)
```

```python
del df["index"]
```

```python
df["coordinates"] = df.apply(lambda row: f"{row['lat']},{row['lon']}",axis=1)
```
